```python
import pandas as pd
import numpy as np

def create_dataframes(data, label):
    dataframes = []
    
    for i in range(45):
        data_i = data[i]
        label_i = label[i]
        data_with_label_i = np.concatenate((data_i, label_i[:, np.newaxis]), axis=1)
        df_i = pd.DataFrame(data_with_label_i)
        dataframes.append(df_i)
    
    return dataframes

# Usage
dataframes = create_dataframes(data, label)
```
